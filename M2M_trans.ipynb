```python
import assemblyai as aai
import os
import json

# Set your API key
aai.settings.api_key = "295cbc3589c7408ca09232c6642f531d"

# Specify the single audio file path (update this with your file path)
audio_file = "/content/sample_data/Doctor's_Hindi.mp3"  # Change this to your file path

# Ensure the file exists before proceeding
if not os.path.exists(audio_file):
    raise FileNotFoundError(f"File not found: {audio_file}")

# Define the transcription config
config = aai.TranscriptionConfig(
    speaker_labels=True,
    #speakers_expected=2,
    language_code="hi"
)

# Transcribe the single audio file
transcript = aai.Transcriber().transcribe(audio_file, config)

# Store transcript in dictionary
transcriptions_dict = {
    os.path.basename(audio_file): [
        {"speaker": utterance.speaker, "text": utterance.text}
        for utterance in transcript.utterances
    ]
}

# Print the transcription with speaker labels
print(f"Transcription for {audio_file}:\n")
for utterance in transcript.utterances:
    print(f"Speaker {utterance.speaker}: {utterance.text}")

print("\n" + "-" * 50 + "\n")

# Save the transcription to a JSON file
output_file = "transcription.json"
with open(output_file, "w", encoding="utf-8") as f:
    json.dump(transcriptions_dict, f, indent=4, ensure_ascii=False)

print(f"Transcription saved to {output_file}")

# Read the JSON file
with open(output_file, "r", encoding="utf-8") as f:
    transcription_data = json.load(f)

# Convert transcription into formatted conversation text
conversation_text = ""
for utt in transcription_data[os.path.basename(audio_file)]:
    conversation_text += f"Speaker {utt['speaker']}: {utt['text']}\n"
```

```python

from transformers import M2M100ForConditionalGeneration, M2M100Tokenizer

hi_text = "जीवन एक चॉकलेट बॉक्स की तरह है।"


model = M2M100ForConditionalGeneration.from_pretrained("facebook/m2m100_418M")
tokenizer = M2M100Tokenizer.from_pretrained("facebook/m2m100_418M")

# translate Hindi to French
tokenizer.src_lang = "hi"
encoded_hi = tokenizer(hi_text, return_tensors="pt")
generated_tokens = model.generate(**encoded_hi, forced_bos_token_id=tokenizer.get_lang_id("fr"))
tokenizer.batch_decode(generated_tokens, skip_special_tokens=True)
```
